```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

from matplotlib import font_manager, rc
font_path = "data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
#1
df = pd.DataFrame({"1분기":[500,690,1100,1500,1990,1020],"2분기":[450,700,1030,1650,2020,1600],"3분기":[520,820,1200,1700,2300,2200],"4분기":[610,900,1380,1850,2420,2550]}
                  ,index=[2015,2016,2017,2018,2019,2020])
display(df)
```

```python
#2
plt.figure(figsize=(10,6))
df.index = ["2015년","2016년","2017년","2018년","2019년","2020년"]
plt.plot(df.index,df['1분기'],
         df.index,df['2분기'],
         df.index,df['3분기'],
         df.index,df['4분기'],)
plt.legend(df.columns)
plt.title('시각화 과제1',color = "purple",fontsize=20)
plt.savefig("output/hw1.png") #show() 호출전에
plt.show()
```

```python
#3
plt.figure(figsize=(10,6))
plt.plot(df.columns,df.T['2015년'],
         df.columns,df.T['2016년'],
         df.columns,df.T['2017년'],
         df.columns,df.T['2018년'],
         df.columns,df.T['2019년'],
         df.columns,df.T['2020년'])
plt.legend(df.T.columns)
plt.title("시각화 과제2",color='red',fontsize=15)
plt.savefig("output/hw2.png")
plt.show()
```

```python
#4
plt.figure(figsize=(10,6))
plt.pie(df.sum(),labels = df.columns,autopct='%.1f%%',explode = [0.05, 0.05, 0.05, 0.05],shadow=True,colors=["red","green","brown","purple"])
plt.savefig("output/hw3.png")
plt.show()
```
