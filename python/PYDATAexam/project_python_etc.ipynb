```python
#구별 6년 평균 미세먼지 지도시각화
import json
#geo_path = 'data/seoul_geo.json'
geo_path = 'data/seoul.geojson'
geo_str = json.load(open(geo_path, encoding='utf-8'))
#print(geo_str)
```

```python
misegu = pd.read_excel('data/미세먼지평균량데이터.xlsx')
display(misegu.T.head())
```

```python
misegu = misegu.T
```

```python
misegu.columns=misegu.loc['year']
```

```python
misegu.drop('year',axis=0,inplace = True)
```

```python
misegu.reset_index(inplace=True)
```

```python
misegu
```

```python
#pop = pd.read_csv("./data/cctv_seoul.csv")
pop = misegu

df = pd.read_excel('./data/미세먼지 측정소별 교통량.xlsx')
df = df[['미세먼지 측정소 명','도로명 주소','위도,경도']]
df['위도,경도'] = df['위도,경도'].replace(" ","")
latf = []
for i in df['위도,경도']:
    latf.append(i.split(','))
latlng = latf
lat = []
lng = []
for i in range(len(latlng)):
    lat.append(latlng[i][0])
    lng.append(latlng[i][1])
df['lat'] = lat
df['lng'] = lng


map = folium.Map(location=[37.5502, 126.982], zoom_start=11.4, tiles =None
                 )
fmap=folium.Choropleth(geo_data = geo_str,
               data = pop,
               columns = ['index', '6년평균'],
               fill_color = 'YlGnBu',                
               key_on='feature.properties.name').add_to(map)
fmap.geojson.zoom_on_click = False
fmap.geojson.add_child(
    folium.features.GeoJsonTooltip(['name'],labels=False)
)

for name, lat, lng in zip(df['미세먼지 측정소 명'], df.lat,df.lng):
    folium.CircleMarker([lat,lng],
                        radius=4,         # 원의 반지름
                        color='brown',         # 원의 둘레 색상
                        fill=True,
                        fill_color='orange',    # 원을 채우는 색
                        fill_opacity=0.7, # 투명도    
                        tooltip=name
    ).add_to(map)


display(map)
map.save('output/misegu.html')
```
