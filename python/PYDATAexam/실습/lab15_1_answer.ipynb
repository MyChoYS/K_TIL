```python
import pandas as pd
import seaborn as sns
import numpy as np
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
df = iris_x
```

```python
#1 (1). 누락데이터가 포함된 행 모두 삭제

df1 = df.dropna()   #기본은 axis=0(행)
display(df1.head(3))
```

```python
#1 (2). 누락데이터가 2개 이상인 행 모두 삭제

df2 = df.dropna(thresh=2)
display(df2.head(10))
```

```python
#1 (3). sepal_length와 sepal_width 변수에서 누락데이터가 있는 행 모두 삭제

df3 = df.dropna(subset = ['sepal_length', 'sepal_width'])   #기본값은 how='any'
display(df3.head(10))
```

```python
#1 (4). 누락데이터를 0으로 치환

df4 = df.fillna(0)
display(df4.head(10))
```

```python
#1 (5). 누락데이터를 평균값으로 치환

df5 = df.fillna(round(df.mean(), 1))
display(df5.head(10))
```

```python
#1 (6). 누락데이터를 10으로 치환

df6 = df.fillna(10)
display(df6.head(10))
```

```python
#1 (7). 누락데이터를 이웃하는 값으로 치환

df7 = df.fillna(method = 'ffill')
display(df7.head(10))
```

```python
#2 (1). 오늘부터 전공 프로젝트 발표 날짜까지의 Timestamp 객체들로 구성된 DateTimeIndex 객체

ts = pd.date_range('2021-04-05', '2021-04-27')  #키워드 생략 가능, 각각 start와 end. freq기본값은 freq='D'
print(ts)
```

```python
#2 (2). 오늘부터 61일 기간의 Timestamp 객체들로 구성된 DateTimeIndex 객체

ts = pd.date_range('2021-04-05', periods = 61)
print(ts)
```

```python
#2 (3). 1월부터 12월까지의 Timestamp 객체들로 구성된 DateTimeIndex 객체(마지막 날짜 기준)

ts = pd.date_range('2021-01-01',
                    periods = 12,
                    freq = 'M')
print(ts)
```

```python
#2 (4). 1월부터 12월까지의 Timestamp 객체들로 구성된 DateTimeIndex 객체(시작 날짜 기준)

ts = pd.date_range('2021-01-01',
                    periods = 12,
                    freq = 'MS')
print(ts)
```

```python
#2 (5). 4월 5일부터 9일까지 시간 단위로 구성되는 Timestamp 객체들로 구성된 DateTimeIndex 객체

ts = pd.date_range('2021-04-05',
                   '2021-04-10',
                   freq = 'H',
                   closed = 'left')
print(ts)
```

```python
#2 (6). 4월 1일부터 4월 27일까지 3시간 단위로 구성되는 Timestamp 객체들로 구성된 DateTimeIndex 객체

ts = pd.date_range('2021-04-01',
                   '2021-04-28',
                   freq = '3H',
                   closed = 'left')
print(ts)
```

```python
#2 (7). 1월 1일부터 두 달 간격으로 8개 기간으로 구성되는 Timestamp 객체들로 구성된 DateTimeIndex 객체

ts = pd.date_range('2021-01-01',
                   periods = 8,
                   freq = '2M')
print(ts)
```

```python
#3 (1).

df = pd.read_csv('./data/product_click_new.log', header=None, sep=' ')

print(df.head(5))
print(df.info())
```

```python
#3. (2)

df.columns = ['oldclicktime', 'clickpid']
print(df.head(5))
```

```python
#3. (3)

df['oldclicktime'] = df['oldclicktime'].astype(str)
print(df.info())
```

```python
#3. (4)

df['newclicktime'] = pd.to_datetime(df['oldclicktime'])
print(df.info())
print(df.head(5))
```

```python
#3. (5)

df['year'] = df['newclicktime'].dt.year
df['month'] = df['newclicktime'].dt.month
df['day'] = df['newclicktime'].dt.day
df['hour'] = df['newclicktime'].dt.hour
df['minute'] = df['newclicktime'].dt.minute

display(df.head(5))
```
