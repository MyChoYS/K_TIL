```python
import requests
from bs4 import BeautifulSoup
import re

r = requests.get('https://comic.naver.com/genre/bestChallenge.nhn')
r.encoding = 'utf-8' 

bs = BeautifulSoup(r.text, 'html.parser')

titlelist = []
summarylist = []
ratinglist = []




for i in range(1,10) :
    title = bs.select(' div.challengeInfo > h6 > a')
    summary = bs.select(' div.challengeInfo > div.summary')
    rating =  bs.select(' div.challengeInfo > div.rating_type > strong')

    for i in title:
        i = re.sub("[\n\t]", '', i.text) 
        i = re.sub(r"^\s+", '', i) #시작 공백 제거
        i = re.sub(r"\s+$", '', i) #끝 공백 제거 
        titlelist.append(i)
    
    for i in summary:
        i = re.sub("[\n\t]", '', i.text)
        summarylist.append(i) 

    for i in rating:
        i = re.sub("[\n\t]", '', i.text)
        ratinglist.append(i)


print(titlelist) 
print("------------------------")
print(summarylist) 
print("------------------------")
print(ratinglist) 
  
    
```
