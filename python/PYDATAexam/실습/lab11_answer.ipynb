```python
# 1 
import pandas as pd

file_path = './data/score.csv'

df = pd.read_csv(file_path)
print("컬럼명 :" , df.columns)
print("인덱스 :" , df.index)

df['total'] = df['kor'] + df['eng'] + df['mat']
df['avg'] = round(df['total'] /3, 2)
display(df)
```

```python
# 2
import pandas as pd

file_path = './data/score_noheader.csv'

df = pd.read_csv(file_path, header=None)
display(df)
print("컬럼명 :" , df.columns)
print("인덱스 :" , df.index)
print("컬럼명 추가 ----------")

df.columns = ['name', 'kor', 'eng', 'mat']
print("컬럼명 :" , df.columns)
display(df)

print("total과 avg열 추가 ----------")
df['total'] = df['kor'] + df['eng'] + df['mat']
df['avg'] = round((df['kor'] + df['eng'] + df['mat'])/3, 2)
display(df)
```

```python
# 3
import pandas as pd

file_path = './data/score_header.csv'

df = pd.read_csv(file_path, header=3)
display(df)
print("컬럼명 :" , df.columns)
print("인덱스 :" , df.index)

df['total'] = df['kor'] + df['eng'] + df['mat']
df['avg'] = round((df['kor'] + df['eng'] + df['mat'])/3, 2)
display(df)

df.to_csv("./output/score_result.csv", mode='a', index=False, header=False)
```
