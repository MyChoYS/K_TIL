```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

r = requests.get('http://media.daum.net/ranking/popular/')
r.encoding = 'utf-8' 

bs = BeautifulSoup(r.text, 'html.parser')
newstitle = []
newscomname = []


title = bs.select('#mArticle > div.rank_news > ul.list_news2 > li > div.cont_thumb > strong > a')
com = bs.select('#mArticle > div.rank_news > ul.list_news2 > li > div.cont_thumb > strong > span')

for i in title:
    i = re.sub(",", '', i.text) #','가 들어가있으면 csv로 저장할 때 구분자로 인식햄 
    newstitle.append(i)
    
for i in com:
    i = re.sub(",", '', i.text)
    newscomname.append(i)
```

```python
df = pd.DataFrame(newscomname,newstitle)
df = df.reset_index()
df.columns = ["newstitle",'newscomname']
print(df)
```

```python
df.to_csv("../output/news_new.csv",index=False)
```
