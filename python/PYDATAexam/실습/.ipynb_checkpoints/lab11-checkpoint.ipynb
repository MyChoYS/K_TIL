```python
import pandas as pd
```

```python
#1
df1 = pd.read_csv("./data/score.csv")
print("컬럼명 :",df1.columns)
print("인덱스 :",df1.index)
df1['total'] = df1['kor'] + df1['eng'] + df1['mat']
df1['avg'] = round((df1['kor'] + df1['eng'] + df1['mat'])/3,2)

display(df1)


```

```python
#2
df2 = pd.read_csv("./data/score_noheader.csv",header = None)
display(df2)
print("컬럼명 :",df2.columns)
print("인덱스 :",df2.index)
print("컬럼명 추가 ------------")
df2.columns = ["name","kor","eng","mat"]
print("컬럼명 :",df2.columns)
display(df2)
print("toatl과 avg열 추가 ---------")
df2['total'] = df2['kor'] + df2['eng'] + df2['mat']
df2['avg'] = round((df2['kor'] + df2['eng'] + df2['mat'])/3,2)
display(df2)
```

```python
#3
df3 = pd.read_csv("./data/score_header.csv",header=3)
display(df3)
print("컬럼명 :",df3.columns)
print("인덱스 :",df3.index)
df3['total'] = df3['kor'] + df3['eng'] + df3['mat']
df3['avg'] = round((df2['kor'] + df3['eng'] + df3['mat'])/3,2)
display(df3)

df3.to_csv("./output/score_result.csv",index=False,header=False)
```
