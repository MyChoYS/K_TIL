```python
import pandas as pd
import seaborn as sns
import numpy as np
import datetime
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
#1-1
df1 = iris_x.head(10)
df1 = df1.dropna(how='any', axis=0)
display(df1)
```

```python
#1-2
df2 = iris_x
df2 = df2.dropna(thresh=3, axis=0)
display(df2.head(10))
```

```python
#1-3
df3 = iris_x
df3 = df3.dropna(subset=['sepal_length','sepal_width'], how='any', axis=0)  
display(df3.head(10))
```

```python
#1-4
df4 = iris_x
df4 = df4.fillna(0)
display(df4.head(10))
```

```python
#1-5
df5 = iris_x
df5['sepal_length'] = df5['sepal_length'].fillna(5.8)
df5['sepal_width'] = df5['sepal_width'].fillna(3.1)
df5['petal_width'] = df5['petal_width'].fillna(1.2)
df5['petal_length'] = df5['petal_length'].fillna(3.7)
display(df5.head(10))
```

```python
#1-6
df6 =iris_x
df6 = df6.fillna(10)
display(df6.head(10))
```

```python
#1-7
df7 = iris_x
df7 = df7.fillna(df7.median(axis=0))
display(round(df7.head(10),1))
```

```python
#3-1
product = pd.read_csv('data/product_click_new.log',header=None,sep=" ")
print(product.head(5))
print(product.info())
print(product[0])
```

```python
#3-2
product.columns = ['oldclicktime','clickpid']
print(product.head(5))
```

```python
#3-3
print(product.info())
```

```python
#3-4
product['newclicktime'] = pd.to_datetime(product['oldclicktime'],format='%Y%m%d%H%M')
product['oldclicktime'] = product['oldclicktime'].astype(str)
print(product.info())
display(product.head(5))
```

```python
#3-5
product['year'] = product['newclicktime'].dt.year
product['month'] =product['newclicktime'].dt.month
product['day'] =product['newclicktime'].dt.day
product['hour'] =product['newclicktime'].dt.hour
product['minute'] =product['newclicktime'].dt.minute
print(product.info())
print(product.head(8))
```
