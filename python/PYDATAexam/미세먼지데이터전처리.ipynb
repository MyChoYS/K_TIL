```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.preprocessing import LabelBinarizer
```

```python
munzi = pd.read_excel('data/측정소없는미세먼지모델.xlsx')
display(munzi.head())
```

```python
years_encoder = LabelBinarizer()
mise_year = munzi['연도']
```

```python
years_encoder.fit(mise_year)
transformed = years_encoder.transform(mise_year)
transformed
```

```python
years_encoder.classes_
```

```python
year_df = pd.DataFrame(transformed, columns=years_encoder.classes_)
year_df
```

```python
munzis = pd.concat([munzi,year_df], axis=1) #열병합
munzis.drop('연도',axis=1,inplace=True)
display(munzis.head())
```

```python
munzis = munzis[[2015,2016,2017,2018,2019,2020,'미세먼지','소매/음식/생활서비스','소매(범위)','음식(범위)','생활서비스(범위)','교통량(양방)','1km내 복원하천 여부','녹지면적 합계']]
```
