```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
```

```python
from matplotlib import font_manager, rc
font_path = "data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
print(font_name)
rc('font', family=font_name)
```

```python
doginfo = pd.read_csv('data/유기동물현황.csv',encoding='ANSI')
doginfo = doginfo[['기간','자치구','계(개)','인도(주인)','입양분양','폐사안락사','계류기증']]
print(doginfo)
```

```python
print(doginfo[doginfo['자치구']=='합계'])
```

```python
seouldog = doginfo[doginfo['자치구']=='합계']
print(seouldog)
```

```python
seouldog['계(개)'] = seouldog['계(개)'].str.replace(',','')
seouldog['계(개)'] = seouldog['계(개)'].astype('int')
print(seouldog.info())
```

```python
seouldog['기간'] = seouldog['기간'].astype('str')
```

```python
plt.figure()
#plt.style.use('default')
plt.figure(figsize=(14,9))
plt.plot(seouldog['기간'],seouldog['계(개)'], linestyle='dashed',color='orange')
plt.title('서울시 유기견 현황',fontsize=25,pad=20)
plt.xlabel('연도',fontsize=25,)
plt.ylabel('유기견 수',fontsize=25)
plt.grid()
plt.savefig("output/서울시 유기견 현황.png")
plt.show()
```

```python
plt.figure(figsize=(15,9))
#plt.style.use(‘default’)
plt.plot(seouldog['기간'],seouldog['계(개)'], linestyle='dashed',color='green')
plt.title('서울시 유기견 현황',fontsize=25)
plt.xlabel('연도',fontsize=25)
plt.ylabel('유기견 수',fontsize=25)
plt.grid()
plt.show()
```
